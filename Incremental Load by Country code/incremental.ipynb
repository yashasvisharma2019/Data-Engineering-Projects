```python

df = spark.read \
    .option("header", "true") \
    .option("delimiter", "\t") \
    .csv("abfss://raw12@datalakestorage1332.dfs.core.windows.net/population_by_age.tsv")
display(df)


```

```python
new_column_names = ["time", "year2008", "year2009", "year2010", "year2011", "year2012","year2013", "year2014", "year2015","year2016","year2017","year2018","year2019"]
df = df.toDF(*new_column_names)
```

```python
from pyspark.sql.functions import *
df_grp = df.withColumn('age_group',regexp_replace(split(df['time'],',')[0],'PC_','')).withColumn('country_code',split(df['time'],',')[1])
df_grp.show(20)
```

```python
df_grp.createOrReplaceTempView("population_by_age")
df_clean = spark.sql("select country_code, age_group, CAST(REGEXP_REPLACE(year2019, '[a-z]', '') AS DECIMAL(4,2)) AS year2019,\
    CAST(REGEXP_REPLACE(year2018, '[a-z]', '') AS DECIMAL(4,2)) AS year2018,\
            CAST(REGEXP_REPLACE(year2017, '[a-z]', '') AS DECIMAL(4,2)) AS year2017,\
                CAST(REGEXP_REPLACE(year2016, '[a-z]', '') AS DECIMAL(4,2)) AS year2016,\
                    CAST(REGEXP_REPLACE(year2015, '[a-z]', '') AS DECIMAL(4,2)) AS year2015,\
                        CAST(REGEXP_REPLACE(year2014, '[a-z]', '') AS DECIMAL(4,2)) AS year2014,\
                            CAST(REGEXP_REPLACE(year2013, '[a-z]', '') AS DECIMAL(4,2)) AS year2013,\
                                CAST(REGEXP_REPLACE(year2012, '[a-z]', '') AS DECIMAL(4,2)) AS year2012,\
                                    CAST(REGEXP_REPLACE(year2011, '[a-z]', '') AS DECIMAL(4,2)) AS year2011,\
                                        CAST(REGEXP_REPLACE(year2010, '[a-z]', '') AS DECIMAL(4,2)) AS year2010,\
                                            CAST(REGEXP_REPLACE(year2009, '[a-z]', '') AS DECIMAL(4,2)) AS year2009,\
                                            CAST(REGEXP_REPLACE(year2008, '[a-z]', '') AS DECIMAL(4,2)) AS year2008      from population_by_age WHERE LENGTH(country_code) = 2")
df_clean.show()
```

```python

df_clean.write.format("delta").mode("overwrite").partitionBy("country_code").save("abfss://raw12@datalakestorage1332.dfs.core.windows.net/cleaned_data/delta_country_code")
```
